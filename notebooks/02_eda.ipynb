```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/House_Price_Prediction/data/raw/Housing.csv')
df.head()
```

```python
plt.hist(df['price'],bins=30)
plt.xlabel('Price (million)')
plt.ylabel('Frequency')
plt.title('Distribution of Housing Prices')
plt.show()
```

```python
sns.distplot(df['price'])
plt.xlabel('Price (million)')
plt.ylabel('Frequency')
plt.title('Distribution of Housing Prices')
plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.subplot(1,2,2)
plt.boxplot(df['area'],patch_artist=True,boxprops=dict(facecolor='lightblue'))
plt.ylabel('Area (square feet)')
plt.title('Distribution of Housing Area')

plt.subplot(1,2,1)
sns.boxplot(df['price'])
plt.ylabel('Price (million)')
plt.title('Distribution of Housing Prices')
plt.show()
```

```python
sns.scatterplot(x='area',y='price',data=df)
plt.xlabel('Area (square feet)')
plt.ylabel('Price (million)')
plt.title('Relationship between Area and Price')
plt.show()
```
